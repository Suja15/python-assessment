```python
import csv
import logging
import random

print("Welcome to Tamil Matrimony")

file_name="matrimonial.csv"

def Insert():
    ID=random.randint(1,50)
    print("Id",ID)
    U_ID=ID
    Name=input("Enter your name:")
    Age=int(input("Enter your age:"))
    Height=int(input("Enter your height:"))
    Mothertounge=input("Enter your mothertounge:")
    Education=input("Education Details:")
    
    with open(file_name,'a',newline='') as file:
        csvwritter=csv.writer(file)
        csvwritter.writerow([U_ID,Name,Age,Height,Mothertounge,Education])
        
def Retrive():
    print("-"*90)
    print("U_ID\t\tName\t\tAge\t\tHeight\t\tMothertounge\t\tEducation")
    print("-"*90)
    
    with open(file_name,'r') as file:
        data=csv.reader(file)
        for row in list (data):
            print(f"{row[0]}\t\t{row[1]}\t\t{row[2]}\t\t{row[3]}\t\t{row[4]}\t\t{row[5]}")
        
def Search():
    nametosearch=input("Enter the name to search:")
    
    print("Search Results")
    print("-"*90)
    print("U_ID\t\tName\t\tAge\t\tHeight\t\tMothertounge\t\tEducation")
    print("-"*90)
    
    with open(file_name,'r') as file:
        data=csv.reader(file)
        details=list(data)
        searchresult=[detailslist for detailslist in details if nametosearch in detailslist]
        if len(searchresult)>0:
            for row in searchresult:
                print(row[0],"\t\t",row[1],"\t\t",row[2],"\t\t",row[3],"\t\t",row[4],"\t\t",row[5],"\n")
        else:
            print("---------------------------------No details found-------------------------------------")

def Delete():
    nametodelete=input("Enter the name to delete:")
    with open(file_name,'w') as file:
        data=csv.writer(file)
        details=list(data)
        finalresult=[detailslist for detailslist in details if nametodelete != detailslist[0]]
    if len(finalresult)!=len(details):
        print("---------------------------------No data found to delete-------------------------------------")
    else:
        print("---------------------------------Updated details-------------------------------------")
        
def menu():
    print("Menu")
    print("0. Exit")
    print("1. Insertion") 
    print("2. Modification") 
    print("3. Retrieval") 
    print("4. Searching") 
    print("5. Recommendation") 
    print("6. Deletion")
    choice=None
    try:
        choice=eval(input("Enter the choice:"))
    except:
        logging.error("Enter the choice as number:")
    return choice

while True:
    choice=menu()
    if choice==0:
        break
    elif choice==1:
        Insert()
    elif choice==2:
        Modify()
    elif choice==3:
        Retrive()
    elif choice==4:
        Search()
    elif choice==5:
        Recommend()
    elif choice==6:
        Delete()
    else:
        pass
```
